```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator

U_target = Operator.from_label('H')  # Hadamard
U_approx = Operator.from_label('X')  # just for demo

error = np.linalg.norm(U_target.data - U_approx.data)
print("Error:", error)
```

```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator

H_operator = Operator.from_label('H')  # Hadamard
H = H_operator.data

theta = np.pi / 4
Rz = np.array([
    [np.exp(-1j * theta / 2), 0],
    [0, np.exp(1j * theta / 2)]
])
state = np.array([[1], [0]])
new_state = H @ Rz @ Rz @ Rz @ Rz @ H @ state
prob0 = np.abs(new_state[0, 0])**2
prob1 = np.abs(new_state[1, 0])**2

print(new_state)
print(prob0, prob1)
```
